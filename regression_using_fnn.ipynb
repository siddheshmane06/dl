```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
x = np.random.rand(1000,3) #vectors will be generated
x[0:5]
```

```python
def true_func(x):
  return np.sin(1.5 * np.pi * x[:,0]) + np.cos(1.5 * np.pi*x[:,1]) + 2 * x[:,2]
```

```python
y = true_func(x)
```

```python
y[0:5]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model = Sequential()
model.add(Dense(15, input_dim=3, activation="relu"))
model.add(Dense(1, activation="linear"))
model.compile(loss="mean_squared_error", optimizer="adam", metrics="mean_absolute_error")
model.fit(xtrain, ytrain, epochs=150, batch_size=30)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest, predictions)
```

### Make a dataset with 3 columns : mass, acceleration and force, where mass and acceleration will be input features and force will be the target to be predicted. Design a neural network which takes mass and acceleration as input and predicts the corresponding force. (Force = Mass * Acceleration)

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0, 5, size=(20, 3)), columns=['mass', 'acceleration', 'force'])
df["force"] = df["mass"]*df["acceleration"]
print(df)
```

```python
x = df.iloc[:,:2]
y = df.iloc[:,2]
print("x:",x)
print("y:",y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 1)
y_train
```

```python
model = Sequential()
model.add(Dense(15, input_dim=2, activation="relu"))
model.add(Dense(1, activation="linear"))
model.compile(loss="mean_squared_error", optimizer="adam", metrics="mean_absolute_error")
model.fit(x_train, y_train, epochs=1000, batch_size=5)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, predictions)
```
